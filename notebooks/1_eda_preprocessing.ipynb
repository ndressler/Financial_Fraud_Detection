## Dimensionality Reduction on Missing Values

Columns with more than 75% missing data are often dropped because they contain too little information for reliable learning. The high proportion of missing values can lead to bias, overfitting, or poor performance in your machine learning models. Furthermore, imputing such a large amount of missing data can introduce additional bias or distort the underlying data distribution. Therefore, it's generally more efficient and effective to drop these columns and focus on the data you do have.

```python
# drop columns containing 75% or more of values missing
old_X = X.shape
old_test = test.shape

X, test = prep.drop_na_cols(X, test)

print(f'Old vs new size of train: {old_X} vs {X.shape}')
print(f'Old vs new size of test: {old_test} vs {test.shape}')
```

## Handle Infinite Values

Handling infinite values and negative infinite values is crucial in data preprocessing because many machine learning algorithms cannot handle these values and will produce errors or incorrect results if they are present in the data.

```python
# replace inf and -inf with NaN
X_inf = X.replace([np.inf, -np.inf], np.nan)
test_inf = test.replace([np.inf, -np.inf], np.nan)
```

Label encoding is a preprocessing step that transforms categorical data into a numerical format, allowing machine learning algorithms to process it. This is particularly useful when dealing with categorical features that have a large number of unique values, as it can efficiently convert these into numerical labels. Furthermore, certain machine learning models, like tree-based models, can handle label-encoded data without assuming an ordinal relationship, making label encoding a suitable choice.

Must be executed before handling missing values since Simple Imputer can only have numerical features.

```python
# columns to encode
categorical_cols_X = X_inf.select_dtypes(include=['object']).columns.tolist()
categorical_cols_test = test_inf.select_dtypes(include=['object']).columns.tolist()

# fit and transform X
X_enc = X_inf

# store the fitted encoders to ensure consistency in encoding
encoders = {}

for col in categorical_cols_X:
    le = LabelEncoder()
    # Add 'Unknown' and 'nan' to the set of unique categories from X_inf and test_inf
    unique_categories = pd.concat([X_inf[col].dropna(), test_inf[col].dropna(), pd.Series(['Unknown', 'nan'])]).unique()
    le.fit(unique_categories.astype(str))
    X_enc[col] = le.transform(X_inf[col].astype(str).fillna('Unknown'))
    encoders[col] = le

# transform test using the fitted encoders
test_enc = test_inf
for col in categorical_cols_X:
    le = encoders[col]
    test_enc[col] = le.transform(test_inf[col].astype(str).fillna('Unknown'))
```

## Handle Missing Values


Many machine learning algorithms cannot handle missing values and will produce an error if we try to fit them with a dataset that contains NaN values. SimpleImputer provides different strategies for imputing missing values, such as mean, median, most_frequent, and constant. In this case, the mean strategy is used, which replaces each missing value in a column with the mean value of that column. This is a simple, fast, and often effective way to handle missing values.

```python
# Simple imputation
simple_imputer = SimpleImputer(strategy='mean')

X_simp = simple_imputer.fit_transform(X_enc)
test_simp = simple_imputer.transform(test_enc)
```

```python
# check if there are no more missing values
print(np.isnan(X_simp).sum())
print(np.isnan(test_simp).sum())
```

## Normalizing

Normalization is a preprocessing step that scales numerical features to a standard range, typically between 0 and 1. This is performed after label encoding to ensure that all features have a similar scale. By normalizing the data, we ensure that each feature contributes approximately proportionately to the final decision of the learning algorithm, preventing features with larger scales from dominating.

The mean and standard deviation of the test data are not the same as those of the training data. This is because the MinMaxScaler was fitted on the training data, not the test data. The same scaling parameters (minimum and maximum of the training data) are used to scale the test data. This prevents information from the test data leaking into the models.

```python
# scaling data using Min Max Scaler
scaler = MinMaxScaler()

X_norm = scaler.fit_transform(X_simp)
test_norm = scaler.transform(test_simp)
```

```python
print("X:")
print("Mean:", X_norm.mean())
print("Standard Deviation:", X_norm.std())

print("Test:")
print("Mean:", test_norm.mean())
print("Standard Deviation:", test_norm.std())
```

## Feature Engineering

creates new features that are ratios of existing features to their group-wise mean and standard deviation, For each combination of feature and group, it creates two new features:

The first new feature is the ratio of the original feature to its mean grouped by the current group.
The second new feature is the ratio of the original feature to its standard deviation grouped by the current group.

```python
# convert the data back to dataframes
X_fe = pd.DataFrame(X_norm, columns=X_cols)
test_fe = pd.DataFrame(test_norm, columns=test_cols)

# aggregate features
X_fe = prep.create_features(X_fe)
test_fe = prep.create_features(test_fe)

# save column names
new_X_cols = X_fe.columns.tolist()
new_test_cols = test_fe.columns.tolist()
print( set(new_X_cols) == set(new_test_cols))
```

```python
# Check if there are any NaN values in the data
print(X_fe.isna().any().any())
print(test_fe.isna().any().any())
```

## Feature Selection

```python
# Define the selector
selector = VarianceThreshold(threshold=0.05)

# Determine the number of chunks
chunk_size = 10000
num_chunks = int(np.ceil(X_fe.shape[0] / chunk_size))

# Initialize an empty list to hold the selected features
selected_features = []

# Process each chunk
for i in range(num_chunks):
    start = i * chunk_size
    end = min((i + 1) * chunk_size, X_fe.shape[0])
    X_chunk = X_fe.iloc[start:end, :]

    # Fit the selector to the chunk
    X_fs = selector.fit_transform(X_chunk)

    # Get the selected features for this chunk
    chunk_selected_features = X_chunk.columns[selector.get_support(indices=True)]

    # Add the selected features to the list
    selected_features.extend(chunk_selected_features)

# Remove duplicates from the list of selected features
selected_features = list(set(selected_features))

# Print the number of features
print("Original number of features: ", X_fe.shape[1])
print("Number of features after variance threshold: ", len(selected_features))
print("Number of features removed: ", X_fe.shape[1] - len(selected_features))
```

```python
X_fs = X_fe[selected_features]
test_fs = test_fe[selected_features]
```

```python
# check features
X_fs_cols = X_fs.columns.tolist()
test_fs_cols = test_fs.columns.tolist()

print(set(X_fs_cols) == set(test_fs_cols))
```

## Convert Fraud Values

The algorithms to be used (IF, LOF and OCSVM) all predict the data in -1 (fraud) and 1 (normal), so we will convert the values in the target data to match the algorithms, making 0 (nor fraud) into 1 and 1 (fraud) into -1.

```python
# convert 1 to -1 and 0 to 1
y = y.map(lambda i: -1 if i == 1 else 1)
```

## Split Train, Validation and Test Set

Here we are splitting the training data into a smaller training set and a validation set allows you to evaluate and tune the models during development without using the test set, providing an unbiased estimate of how the model will perform on unseen data.

0.6 for train<br>
0.2 for test<br>
0.2 for validation

```python
# split the data into a temporary training set and a test set
X_temp, X_test, y_temp, y_test = train_test_split(X_fs, y, test_size=0.2, random_state=42, stratify=y)

# split the temporary training set into a final training set and a validation set
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=42, stratify=y_temp)  # 0.25 x 0.8 = 0.2

# convert arrays back to DataFrames
X_train = pd.DataFrame(X_train, columns=X_fs_cols)
X_val = pd.DataFrame(X_val, columns=X_fs_cols)
X_test = pd.DataFrame(X_test, columns=X_fs_cols)

y_train = pd.DataFrame(y_train, columns=['isFraud'])
y_val = pd.DataFrame(y_val, columns=['isFraud'])
y_test = pd.DataFrame(y_test, columns=['isFraud'])

# rename final test dataset
unlabeled_test = test_fs
```
